## 데이터 전처리

```python
# 발생시간 칼럼으로 담아주기
df['발생시간'] = df['발생년월일시'].str.split(' ').str[1]
# return 이 없다면 잘 적용된 것
```

```python
df.head() # 적용이 잘 되었는지 확인
```

```python
df['발생시간'] = df['발생년월일시'].str[-5:-3]
```

```python
# 3) 발생시간 컬럼을 숫자 데이터 타입으로 변경
#df['발생시간'] = df['발생시간'].str.split(':').str[0]
df = df.astype({'발생시간':'int'})
df.info() # 발생시간 int로 형변환 되었다는 것을 확인
```

```python
# 발생년월일시를 가지고 발생년월일을 날짜타입으로 하여 파생 변수 만들기
df['발생년월일'] = pd.to_datetime(df['발생년월일시'].str[:10], errors='raise')
df.info()
```

```python
df.head()
```

```python
# 발생년월일시 칼럼 지우기
del df['발생년월일시']
```

```python
df.head()
```

```python
# 칼럼 순서 정리

df = df[['발생년'] + df.columns[-2:].tolist()[::-1] + df.columns[1:-2].tolist()]
```

```python
df.columns
```

```python
df.info()
```

```python
# 대전광역시 데이터 준비
dj_df = df.loc[df['발생지시도'] == '대전']
dj_df.info()
```

 Plotly 그래프의 레이아웃을 업데이트하는 메서드입니다. 이 메서드를 사용하여 그래프의 여러 속성을 설정할 수 있습니다.

```python
# 대전광역시 시간대 별 교통사고 사망자 현황 시각화
fig = px.bar(dj_df, x = '사망자수', y = '발생시간', orientation='h')

# 그래프 타이틀 달아주기
fig.update_layout(
    title='대전광역시 시간대 별 교통사고 사망자 현황'
)

fig.show()
```

대전광역시 시간대 별 교통사고는 17시에 가장 많이 발생했고 오전 11시와 새벽 6시에 한건도 발생하지 않았다.

# 지도를 활용한 교통사고 현황 분석하기
# 사망 사고 위치에 CircleMarker 표출
# CircleMarker 사이즈 사망자수 + 경상자수 설정 -> 클릭하면 popup으로 정보 (사고 유형) 나타나도록 구현
# 지도 시각화는 .html 파일로 저장하여 시연

```python
# 기본 지도 준비

map = folium.Map(location=[36.3504119,127.3845475]) # 경도, 위도
map
```

```python
# 사망자수, 경상자수 칼럼 데이터 타입 -> 실수형으로 변경

dj_df = dj_df.astype({'사망자수' : float, '경상자수' : float})
dj_df.head()
```

```python
# 사망자수 + 경상자수에 따른 CircleMarker 표출

map = folium.Map(location=[36.3504119,127.3845475]) # 경도, 위도

# 2) CircleMarker 지도에 표출 -> for문 이용해서 dj_df 데이터가 있을때까지 CircleMarker 지도에 add

for n in  dj_df.index :
  # CircleMarker size
  cnt = dj_df['사망자수'][n] + dj_df['경상자수'][n]
  # 위도, 경도 추출
  lat = dj_df['위도'][n]
  lng = dj_df['경도'][n]

  # CircleMarker 지도에 add
  folium.CircleMarker([lat, lng], radius= cnt * 10, popup= dj_df['사고유형'][n],
                      color = '#F70000', fill_color = 'F70000').add_to(map)

map
```

```python
map.save('2021_대전교통사고현황.html')
```
