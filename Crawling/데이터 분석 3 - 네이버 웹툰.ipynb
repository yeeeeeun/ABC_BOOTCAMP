```python
# 모든 이미지 주소를 가져와봅시다.
```

```python
page_url = "https://comic.naver.com/webtoon/detail?titleId=769209&no=2"

res = requests.get(page_url)
res
```

```python
html = res.text
len(html)
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
soup.select(".wt_viewer img")
```

```python
tag_list = soup.select(".wt_viewer img")
len(tag_list)
```

```python
tag = tag_list[0]
tag
```

```python
# 한 웹툰 내 모든 이미지의 주소를 print해주세요
idx = 0
for tag in tag_list :
    filename = f"{idx}.jpg"
    print(tag["src"])
    
    idx += 1
```

```python
for idx, tag in enumerate(tag_list, 1) :
    filename = f"{idx}.jpg"
    print(filename)
    
    image_res = requests.get(tag['src'], headers = headers)
    assert image_res.status_code == 200, "서버 응답이 200이 아니예요 !!!"
    
    # 저장
    with open(filename, "wb") as f :
        f.write(image_res.content)
```

```python
# "화산귀환/2화" 폴더 안에 웹툰 이미지 저장하기
```

```python
웹툰명 = "화산귀환"
회차 = "2화"

for idx, tag in enumerate(tag_list, 1) :
    filepath = f"{웹툰명}/{회차}/{idx}.jpg"
    print(filepath)
    
    dir_path = os.path.dirname(filepath)
    
    if not os.path.exists(dir_path) :
        os.makedirs(dir_path) 
    
    image_res = requests.get(tag['src'], headers = headers)
    assert image_res.status_code == 200, "서버 응답이 200이 아니예요 !!!"
    
    # 저장
    with open(filepath, "wb") as f :
        f.write(image_res.content)
```

```python
from tqdm.notebook import tqdm # tqdm

웹툰명 = "화산귀환"
회차 = "2화"

for idx, tag in tqdm(enumerate(tag_list, 1)) :
    filepath = f"{웹툰명}/{회차}/{idx:03d}.jpg"
    print(filepath)
    
    dir_path = os.path.dirname(filepath)
    
    os.makedirs(dir_path, exist_ok = True) 
    
    image_res = requests.get(tag['src'], headers = headers)
    assert image_res.status_code == 200, "서버 응답이 200이 아니예요 !!!"
    
    # 저장
    with open(filepath, "wb") as f :
        f.write(image_res.content)
```
