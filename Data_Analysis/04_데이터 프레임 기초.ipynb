### 1. 조건문을 활용해 파생 변수 만들기

```python
# mpg 데이터에서 category가 2seater인 경우만 추출
mpg_df.query("category == '2seater'")
```

```python
# mpg 데이터에서 category가 minivan인 경우만 추출
mpg_df.query("category == 'minivan'")
```

#### 초과 (>), 미만 (<), 이상 (>=), 이하 (<=) 조건 걸기

```python
# 연식이 2000년 초과인 경우
mpg_df.query("year > 2000") # 숫자는 "" 안에서 처
```

```python
# 연식이 2000년 미만인 경우
mpg_df.query("year < 2000") # 숫자는 "" 안에서 처
```

#### 여러 조건을 충족하는 행 추출하기

```python
# minivan 이거나, 연식이 2000년 이상이 경우 추출
mpg_df.query ("category == 'minivan' | year >= 2000")
```

## 02-5 파생변수 추가하기

### 파생변수 추가하기

파생변수 추가하기
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.assign.html

```python
# total (도시 연비 + 고속도로 연비 / 2) 변수 추가
mpg_df.assign(total = (mpg_df['cty'] + mpg_df['hwy']) / 2) # assign () 사용
```

```python
# total (도시 연비 + 고속도로 연비) 변수 추가
# mean (도시 연비 + 고속도로 연비 / 2) 통합 평균 연비 변수
mpg_df.assign(total = (mpg_df['cty'] + mpg_df['hwy']),
              mean = (mpg_df['cty'] + mpg_df['hwy']) / 2 )
```

```python
mpg_df = mpg_df.assign(total = (mpg_df['cty'] + mpg_df['hwy']),
              mean = (mpg_df['cty'] + mpg_df['hwy']) / 2 ) # 실제로 담아 사용
```

## 02-6 집단별로 요약하기

### 집단별로 요약하기

전체 요약통계량 구하기
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.agg.html

```python
# 통함연비 (total) 의 평균 구하기
mpg_df.agg(mean_total = ('total', 'mean'))
```

#### 집단별로 요약 통계량 구하기

```python
# 차종별로 통합연비 평균 구하기

mpg_df.groupby('category').agg(mean_total = ('total', 'mean')) # total 을 평균 구해라 ('칼럼명','통계기법')
# mpg_df['category'] 불가능
```

```python
# 차종별로 통합연비 평균 구하기

mpg_group_df = mpg_df.groupby('category', as_index = False ).agg(mean_total = ('total', 'mean')) # total 을 평균 구해라 ('칼럼명','통계기법') # 인덱스를 깨서 접근하기 위한 방법
mpg_group_df['category']
```

### 집단별로 다시 집단 나누기

```python
mpg = pd.read_csv('/content/mpg.csv')
mpg
```

```python
# 제조회사별로 분류하고 구동 방식별로 분리 -> 도시 연비의 평균
# F는 Front Wheel : 전륜, R은 Rear Wheel : 후륜, 4 : 사륜
mpg.groupby(['manufacturer', 'drv']).agg(mean_cty = ('cty', 'mean'))
```

```python
# 제조사 아우디에서 구동 방식별 차량의 수

# 1) audi 추출하기
# 2) 구동 방식으로 groupby
# 3) 차량의 빈도 구하기 agg -> count 사용
mpg.query("manufacturer == 'audi'").groupby('drv').agg(n = ('drv', 'count'))
```

```python
mpg.query("manufacturer == 'audi'")['drv'].value_counts() # 위 코드와 결과는 같지만 리턴 형식이 다름
```

```python
# 제조사 volkswagen에서 연식별 차량의 수

# 1) volkswagen 추출하기
# 2) 연식으로 groupby
# 3) 차량의 빈도 구하기 agg -> count 사용
mpg.query("manufacturer == 'volkswagen'").groupby('year').agg(count_year = ('year', 'count'))
```
