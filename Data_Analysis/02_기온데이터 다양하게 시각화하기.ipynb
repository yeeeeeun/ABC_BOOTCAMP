01월 (파란색)과 08월(빨간색)의 최고 기온의 빈도수를 한번에 시각화 하기 -> 히스토그램

```python
# 관측 이래, 서울의 최고 기온 데이터를 시각화
import csv
import matplotlib.pyplot as plt # 라이브러리는 순서대로

# 1) 최고 기온 데이터를 담을 리스트 선언
jan_list = [] # 1월
aug_list = [] # 8월

# 2) 파일을 연다.
f = open('/content/seoul_2023.csv', 'r', encoding = 'euc-kr')

# 3)csv 파일을 읽는다.
data = csv.reader(f, delimiter=',')

# 3-1) 헤더(데이터 정보) 저장하기
header = next(data) # next() 함수는 한줄 읽고, 커서(읽어드릴 위치) 한칸 이동

# 3-2) 최고 기온 result 리스트에 담기
for row in data :
  # 3-1-1 ) 결측치 데이터 처리
  if row[-1] != '' :
    # 01월과 8월 데이터만 담기
    month = row[0].split('-')[1]
    if month == '01' :
      jan_list.append(float(row[-1]))
    if month == '08' :
      aug_list.append(float(row[-1]))

# 4) 파일을 닫는다.
f.close()

# 5) 최고 기온 데이터 시각화
plt.figure(dpi=100) # 속성

plt.title("서울의 8월 최고 기온 히스토그램")
plt.hist(jan_list, bins=100, color ='b', label ='1월의 최고 기온') # bins 추가 -> 더 디테일하게 표현 가능
plt.hist(aug_list, bins=100, color ='r', label ='8월의 최고 기온')
plt.legend()
plt.show()

```

서울의 사용자가 입력한 월 최고, 평균, 최저 기본 히스토그램

```python
# 최저 기온 파랑, 평균 초록, 최고 기온 빨강으로 표시
# 범례를 적용하여 하나의 히스토그램으로 그리기

import csv
import matplotlib.pyplot as plt

choice_month = int(input("원하는 월을 입력하세요."))

# 1) 데이터를 담을 리스트 선언
low_list = [] # 원하는 월의 최저 기온 담을 리스트
mean_list = [] # 원하는 월의 평균 기온 담을 리스트
high_list = [] # 원하는 월의 최고 기온 담을 리스트

# 2) 파일을 연다.
f = open('/content/seoul_2023.csv', 'r', encoding = 'euc-kr')

# 3)csv 파일을 읽는다.
data = csv.reader(f, delimiter=',')

# 3-1) 헤더(데이터 정보) 저장하기
header = next(data) # next() 함수는 한줄 읽고, 커서(읽어드릴 위치) 한칸 이동

# 3-3) 평균 기온 result 리스트에 담기
for row in data :
  low = row[-2]
  mean = row[2]
  high = row[-1]

  if low != '' :
    month = int (row[0].split('-')[1])
    if month == choice_month :
      low_list.append(float(low))
  if mean != '' :
    if month == choice_month :
      mean_list.append(float(mean))
  if high != '' :
    if month == choice_month :
      high_list.append(float(high))


# 4) 파일을 닫는다.
f.close()

# 5) 최고 기온 데이터 시각화
plt.figure(dpi=100) # 속성

plt.title("서울의 %.2d월 최고, 평균, 최저 기본 히스토그램" %choice_month)
plt.hist(high_list, bins=100, color ='r', label ='%.2d월 최고기온' %choice_month)
plt.hist(mean_list, bins=100, color ='g', label ='%.2d월 평균기온' %choice_month)
plt.hist(low_list, bins=100, color ='b', label ='%.2d월 최저기온' %choice_month)
plt.legend()
plt.show()

```

```python
import csv
import matplotlib.pyplot as plt

# 1 - 사용할 변수 먼저 선언하기
high_list = []
avg_list = []
low_list =[]

# 2 - 사용자 월 입력 받기
choice_month = (input("원하는 월을 입력하세요."))
if len (choice_month) == 1 :
  choice_month = '0' + choice_month
# 입력된 숫자의 길이에 따라 처리를 해줌

# 3 - 파일을 연다.
f = open('/content/seoul_2023.csv', 'r', encoding = 'euc-kr')

# 4 - csv 파일을 읽는다.
data = csv.reader(f, delimiter=',')

# 4.1 - 헤더(데이터 정보) 저장하기
header = next(data)

# 4.2 - 사용자가 입력한 월과 일치하는 최고, 최저, 평균 기온 데이터 담기
for row in data :
  # 4.1.1 - 사용자가 입력한 월과 일치하는 기온 데이터
  month =  row[0].split('-')[1]
  if month == choice_month :
    # 4.2.2 - 결측치 처리
    if row[-1] != '' and row[-2] != '' and row[-3] != '' :
      high_list.append(float(row[-1]))
      low_list.append(float(row[-2]))
      avg_list.append(float(row[-3]))

# 5) 파일을 닫는다.
f.close()

# 6) 최고 기온 데이터 시각화
plt.figure(dpi=100) # 속성

plt.title("서울의 %s월 최고, 평균, 최저 기본 히스토그램" %choice_month)
plt.hist(high_list, bins=100, color ='r', label = '%s월의 최고기온' %choice_month)
plt.hist(mean_list, bins=100, color ='g', label = '%s월의 평균기온' %choice_month)
plt.hist(low_list, bins=100, color ='b', label = '%s월의 최저기온'%choice_month)
plt.legend()
plt.show()
```
