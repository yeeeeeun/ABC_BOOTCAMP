# 도매업(Wholesale customers) 데이터셋을 활용한 군집 분석

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## 데이터 준비하기


*   https://archive.ics.uci.edu/dataset/292/wholesale+customers
*   도매업 고객의 각 품목에 대한 연간 지출



```python
uci_path = 'https://archive.ics.uci.edu/ml/machine-learning-databases/\
00292/Wholesale%20customers%20data.csv'

df = pd.read_csv(uci_path)
df
```

## 데이터 확인하기

```python
df.info() # 데이터 정보 확인하
```

```python
# Channel - Horeca (Hotel/Restaurant/CafÃ©) or Retail channel (Nominal)
df.Channel.unique() # 영문에 용이
df['Channel'].unique() # 한글에 용이, 가독성이 좋음
```

```python
df.Region.unique() # Lisnon, Oporto or Other (Nominal)
```

## 데이터 전처리

```python
X = df.copy() # 복사, 원본 데이터 유지 위해
X.shape
```

```python
from sklearn import preprocessing
# 데이터 정규화 -> 군집 분석에서는 정규화 중요 -> 데이터 범위가 넓으면 평균 내기가 어려움 -> 범위 축소가 핵심 !
# 범위 축소 (군집간의 거리 계산을 위해)

X = preprocessing.StandardScaler().fit(X).transform(X)
X
```

## K-Means 군집 모델 설정

```python
from sklearn import cluster
import warnings
warnings.filterwarnings('ignore')

kmeans = cluster.KMeans(n_clusters = 5, random_state=7)
```

## 모델 학습하기

```python
kmeans.fit(X)
```

```python
# n_clusters = 5 - > 0 ~ 4 군집 만들기

cluster_label = kmeans.labels_
cluster_label
```

```python
df['Cluster'] = cluster_label
df
```

## 클러스터링 결과 분석 (시각화)

### 지역과 채널과의 연관관계

```python
df.info()
```

```python
# Channel(1,2)
# Region(1,2,3)
# 1 cluster -> 1번 channel ((Hotel/Restaurant/CafÃ©)) 의 3번 region ()
# 2 cluster -> 1번 channel의 3번 region
# 3 cluster -> 1번 channel의 3번 region

df.plot(kind='scatter', x='Channel', y ='Region', c = 'Cluster', cmap = 'Set1', figsize=(10,10))
```

### 품목들간의 연관 관계 ('Fresh', 'Milk')

```python
df.plot(kind='scatter', x='Milk', y ='Fresh', c = 'Cluster', cmap = 'Set1', figsize=(10,10))
```

### 품목들간의 연관 관계 ('Frozen', 'Detergents_Paper')

```python
df.plot(kind='scatter', x='Frozen', y ='Detergents_Paper', c = 'Cluster', cmap = 'Set1', figsize=(10,10))
```

```python
# 특정 클러스터 확인 (2, 3번 클러스터 자세히)
mask = (df['Cluster'] == 2) | (df['Cluster'] == 3)
ndf = df[mask]
ndf['Cluster'].value_counts()
ndf.plot(kind='scatter', x='Frozen', y ='Detergents_Paper', c = 'Cluster', cmap = 'Set1', figsize=(10,10))
```

```python
# 특정 클러스터 확인 (0, 1, 4번 클러스터 자세히)
mask = (df['Cluster'] == 2) | (df['Cluster'] == 3)
ndf = df[~mask]
ndf['Cluster'].value_counts()
ndf.plot(kind='scatter', x='Frozen', y ='Detergents_Paper', c = 'Cluster', cmap = 'Set1', figsize=(10,10))
```

### cluster 비율 확인

```python
df.Cluster.value_counts().plot.pie(autopct='%.2f')
```

## K-Means cluster 3으로 지정하여 모델 수정

```python
kmeans_3 = cluster.KMeans(n_clusters = 3, random_state=7)
kmeans_3.fit(X) # 모델 생성 후 확인
```

```python
ndf = df.copy()
ndf['Cluster'] = kmeans_3.labels_
```

```python
ndf.Cluster.value_counts().plot.pie(autopct='%.2f')
```

```python
ndf.Cluster.value_counts()
```

```python
pd.plotting.scatter_matrix(ndf, c =ndf['Cluster'], figsize=(20,20), marker ='o',
                           alpha=0.8)
plt.tight_layout()
```

```python
import numpy as np

plt.imshow([np.unique(ndf['Cluster'])])
plt.show()
```
