# 문제 정의 : 붓꽃의 품종을 분류


*   3개의 품종 중 하나를 예측하는 다중 분류 문제로 정의
*   문제집 : 데이터, 특성(feature), 독립변수(x) -> 꽃잎, 꽃받침의 길이 (cm) 4가지
*   정답 : 클래스 (class), 레이블(label), 타킷(target), 종속변수(y) -> 붓꽃의 품종 (setosa, versicolor, virginica)



## 데이터 준비하기

```python
from sklearn.datasets import load_iris

iris_dataset = load_iris() # 데이터 셋 불러오기
```

## 데이터 확인하기

```python
iris_dataset['target'] # 정답, label, y
```

```python
iris_dataset['data'] # 문제집, feature, x
```

```python
type(iris_dataset['data']) # 넘파이 array
```

## 데이터와 타킷의 연관성 확인 -> 데이터 분석

scatter chart 그리기 -> 왜 ? 연관성을 확인하기 위해

```python
import matplotlib.pyplot as plt

import pandas as pd
```

```python
# 데이터프레임을 사용하여 데이터 분석 -> data와 target의 연관성 분석

iris_df = pd.DataFrame(iris_dataset['data'], columns=iris_dataset.feature_names)
iris_df.head()
```

```python
iris_df.info()
```

```python
# 각 feature들의 산점도 행렬 4x4

pd.plotting.scatter_matrix(iris_df, c=iris_dataset['target'], figsize=(15,15),
                           marker='o', alpha=.8)
# alpha 는 투명도

plt.show()
```

```python
import numpy as np
```

```python
np.unique(iris_dataset['target']) # 0, 1, 2 가 있다는 것을 알 수 있음 -> 중복 값 제거

plt.imshow([np.unique(iris_dataset['target'])])
_ = plt.xticks(ticks = np.unique(iris_dataset['target']), labels = iris_dataset['target_names'])
# plt.xticks(ticks = np.unique(iris_dataset['target']), labels = iris_dataset['target_names']); 도 같은 역할
```

```python
# 꽃잎(petal) 의 길이와 넓이로 산점도 행렬 그리기 2x2

iris_df2 = iris_df[['petal length (cm)','petal width (cm)']]

pd.plotting.scatter_matrix(iris_df2, c=iris_dataset['target'], figsize=(15,15),
                           marker='o', alpha=.8)

plt.show()
```

## 훈련데이터와 테스트데이터 분리

```python
# 훈련데이터 75%, 테스트데이터 25% -> 랜덤하게 분리
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'],
                                                    test_size= 0.25, random_state=777)
# X 대문자인 이유 - > feature 가 여러개이기 때문.
```

```python
# 훈련 데이터 75% -> 112건의 4개의 feature 로 이루어짐

X_train.shape
```

```python
# 테스트 데이터 25% -> 38건의 4개의 feature 로 이루어짐
X_test.shape
```

```python
y_train
```

## 머신러닝 모델 선택하기 -> K-최근접 이웃 알고리즘

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors= 1)
```

## 학습하기

```python
knn.fit(X_train, y_train) # fit -> 학습 시키기
```

## 예측하기

```python
y_pred = knn.predict(X_test)
```

```python
y_pred
```

## 모델 평가하기

### 정확도 확인하기

```python
# np.mean() 함수를 활용
np.mean(y_pred == y_test)
```

```python
# score() 메서드 활용
knn.score(X_test, y_test)
```

```python
# 평가 지표 계산
from sklearn import metrics

knn_report = metrics.classification_report(y_test, y_pred)
print(knn_report)
```
